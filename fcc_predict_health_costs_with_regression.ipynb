```
# Import libraries. You may or may not use all of these.
!pip install -q git+https://github.com/tensorflow/docs
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

try:
  # %tensorflow_version only exists in Colab.
  %tensorflow_version 2.x
except Exception:
  pass
import tensorflow as tf

from tensorflow import keras
from tensorflow.keras import layers

import tensorflow_docs as tfdocs
import tensorflow_docs.plots
import tensorflow_docs.modeling
```

```
# Import data
!wget https://cdn.freecodecamp.org/project-data/health-costs/insurance.csv
dataset = pd.read_csv('insurance.csv')
dataset.tail()
```

# Task
Build a TensorFlow Keras regression model to predict health insurance expenses with a Mean Absolute Error (MAE) less than 3500.

## Preprocess Data

### Subtask:
Convert categorical features ('sex', 'smoker', 'region') into numerical representations using one-hot encoding. Then, split the dataset into an 80% training set and a 20% test set. Finally, separate the 'expenses' column into `train_labels` and `test_labels`, and create `train_dataset` and `test_dataset` without the 'expenses' column.


**Reasoning**:
The first step is to convert the categorical features ('sex', 'smoker', 'region') into numerical representations using one-hot encoding. This is done using `pd.get_dummies`.



```
dataset_encoded = pd.get_dummies(dataset, columns=['sex', 'smoker', 'region'], drop_first=False)
print("Dataset after one-hot encoding:")
print(dataset_encoded.head())
```

**Reasoning**:
Now that the categorical features are encoded, the next step is to split the `dataset_encoded` DataFrame into an 80% training set and a 20% test set, and then separate the 'expenses' column to create `train_labels`, `test_labels`, `train_dataset`, and `test_dataset`.



```
from sklearn.model_selection import train_test_split

# Split the data into training and testing sets (80% train, 20% test)
train_and_test_data, test_data = train_test_split(dataset_encoded, test_size=0.2, random_state=42)

# Separate 'expenses' column into labels for training and testing sets
train_labels = train_and_test_data['expenses']
test_labels = test_data['expenses']

# Create train_dataset and test_dataset by dropping the 'expenses' column
train_dataset = train_and_test_data.drop('expenses', axis=1)
test_dataset = test_data.drop('expenses', axis=1)

print("Shape of train_dataset:", train_dataset.shape)
print("Shape of train_labels:", train_labels.shape)
print("Shape of test_dataset:", test_dataset.shape)
print("Shape of test_labels:", test_labels.shape)
print("\nFirst 5 rows of train_dataset:\n", train_dataset.head())
print("\nFirst 5 values of train_labels:\n", train_labels.head())
```

## Build and Compile Model

### Subtask:
Create a TensorFlow Keras Sequential model suitable for regression. Define the layers and the output unit. Compile the model using an appropriate optimizer (e.g., Adam), a loss function (e.g., Mean Squared Error), and metrics (e.g., Mean Absolute Error).


**Reasoning**:
The next step is to create the model architecture and compile it as instructed.



```
def build_model():
  model = keras.Sequential([
    keras.Input(shape=[len(train_dataset.keys())]),
    layers.Dense(64, activation='relu'),
    layers.Dense(64, activation='relu'),
    layers.Dense(1)
  ])

  optimizer = tf.keras.optimizers.Adam()

  model.compile(
    loss='mse',
    optimizer=optimizer,
    metrics=['mae', 'mse']
  )
  return model

model = build_model()
model.summary()
```

## Train Model

### Subtask:
Train the compiled model using the `train_dataset` and `train_labels`. Specify the number of epochs and potentially a validation split or callbacks for better training control.


**Reasoning**:
The next step is to train the compiled model using the provided training data and labels, specifying the number of epochs and a validation split.



```
history = model.fit(
    train_dataset,
    train_labels,
    epochs=100,
    validation_split=0.2,
    verbose=0,
    callbacks=[tfdocs.modeling.EpochDots()]
)
```

## Final Task

### Subtask:
Evaluate the trained model's performance and visualize its predictions using the provided test cell. The goal is to achieve a Mean Absolute Error (MAE) less than 3500.


```
# RUN THIS CELL TO TEST YOUR MODEL. DO NOT MODIFY CONTENTS.
# Test model by checking how well the model generalizes using the test set.
loss, mae, mse = model.evaluate(test_dataset, test_labels, verbose=2)

print("Testing set Mean Abs Error: {:5.2f} expenses".format(mae))

if mae < 3500:
  print("You passed the challenge. Great job!")
else:
  print("The Mean Abs Error must be less than 3500. Keep trying.")

# Plot predictions.
test_predictions = model.predict(test_dataset).flatten()

a = plt.axes(aspect='equal')
plt.scatter(test_labels, test_predictions)
plt.xlabel('True values (expenses)')
plt.ylabel('Predictions (expenses)')
lims = [0, 50000]
plt.xlim(lims)
plt.ylim(lims)
_ = plt.plot(lims,lims)
```
